**Importing** **the** **Dependencies**


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**Data** **Collection** **and** **Analysis**

**PIMA** **Diabetes** **Dataset**


```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('ML-diabetes.csv') 
```

```python
# printing the first 10 rows of the dataset
diabetes_dataset.head(10)
```

```python
# number of rows and Columns in this dataset
diabetes_dataset.shape
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> Non-Diabetic

1 --> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and fields.
data = diabetes_dataset.drop(columns = 'Outcome', axis=1)
outcomes = diabetes_dataset['Outcome']
```

```python
print(data)
```

```python
print(outcomes)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
standardized_data = scaler.fit_transform(data)
```

```python
print(standardized_data)
```

```python
data = standardized_data
```

Train Test Split

```python
data_train, data_test, outcomes_train, outcomes_test = train_test_split(data,outcomes, test_size = 0.2, stratify= outcomes, random_state=2)
```

```python
print(data.shape, data_train.shape, data_test.shape)
```

**Training** **the** **Model**

```python
classifier = svm.SVC(C=10)
```

```python
#training the support vector Machine Classifier
classifier.fit(data_train, outcomes_train)
```

**Model** **Evaluation**

Accuracy Score

```python
# accuracy score on the training data
data_train_prediction = classifier.predict(data_train)
training_data_accuracy = accuracy_score(data_train_prediction, outcomes_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
data_test_prediction = classifier.predict(data_test)
test_data_accuracy = accuracy_score(data_test_prediction, outcomes_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

**Making** **a** **Predictive** **system**

```python
input_data = [5,166,72,19,175,25.8,0.587,51]

# changing the input_data to numpy array
numpy_array = np.asarray(input_data)
# reshape the array as we are predicting for one instance
reshaped_array = numpy_array.reshape(1,-1)

# standardize the input data
standard_data = scaler.transform(reshaped_array)
print(standard_data)

prediction = classifier.predict(standard_data)
print(prediction)

if (prediction[0]==0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
